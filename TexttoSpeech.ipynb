# Text-to-Speech (TTS) with gTTS
This notebook demonstrates how to use the Google Text-to-Speech (gTTS) library to convert text into spoken audio. We will generate audio for both English and French text.


Install gTTS: Install the gTTS library to convert text to speech.

English Text-to-Speech:
1. Define the text to be converted and specify the language as English.
2. Create a gTTS object and save the generated speech as an audio file.
3. Play the audio file.

```python
from gtts import gTTS

text_to_say = "This is a samplepiece of audio by GTTS."

language = 'en'


gTTS_object = gTTS(text = text_to_say,
                   lang=language,
                   slow = False)

gTTS_object.save("/content/gTTS.wav")
```

```python
from IPython.display import Audio

Audio("/content/gTTS.wav")
```

French Text-to-Speech:
1. Define French text and specify the language as French.
2. Create a gTTS object, setting slow=True for slower speech, and save the audio file.
3. Play the French audio file.

```python
french_text = "Bonjour. Je m'appelle Alexandra. Je suis programmeuse."

french_language = "fr"

french_ggts_object = gTTS(text = french_text,
                          lang = french_language,
                          slow = True)
french_ggts_object.save("/content/french.wav")
```

```python
from IPython.display import Audio

Audio("/content/french.wav")
```

Code Explanation:

1. The gTTS library is used to convert text to speech.
2. The gTTS object is initialized with text, language, and speed parameters.
3. The generated speech is saved as .wav files and played back using the IPython.display.Audio function.
